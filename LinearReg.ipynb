```
import numpy as np
print(np.__version__)
```

```
x = np.arange(1, 20, 0.5)
y = 2 * x + 1 + np.random.normal(0,2, len(x))
```

```
import matplotlib.pyplot as plt

def plot(x, y, y_pred=None, error=None):
    plt.scatter(x, y, label='Data')
    if y_pred is not None:
        plt.plot(x, y_pred, color='red', label='Prediction')
    plt.xlabel('x')
    plt.ylabel('y')
    if error is not None:
        plt.title(f'Error: {error:.2f}')
    plt.legend()
    plt.show()
```

```
plot(x,y)
```

```
def error(y_pred, y):
    return np.mean((y_pred - y) ** 2)
```

```
col1 = np.ones(38)
X = np.column_stack((col1, x))
Y = y.T

theta = np.linalg.inv(X.T @ X) @ X.T @ Y
y_pred = X @ theta
```

```
plot(x, y, y_pred, error(y_pred, y))
```

```
print(theta)
```

```
x2 = np.arange(1,20,0.5) * 2 + np.random.normal(0,2, len(x))
y2 = 2 * x + 3 * x2 + 2 + np.random.normal(0,2, len(x2))
X2 = np.column_stack((col1, x, x2))

Y2 = y2.T
```

```
def plot_multiple(x, y, y_pred=None, error=None):
    fig = plt.figure(figsize=(12, 4))
    ax1 = fig.add_subplot(131, projection='3d')
    ax1.scatter(x[:, 1], x[:, 2], y, label='Data')
    if y_pred is not None:
        ax1.plot3D(x[:, 1], x[:, 2], y_pred, color='red', label='Prediction')
    ax1.set_xlabel('x1')
    ax1.set_ylabel('x2')
    ax1.set_zlabel('y')
    if error is not None:
        ax1.set_title(f'Error: {error:.2f}')
    ax1.legend()
    plt.show()
```

```
plot_multiple(X2, y2)
```

```
theta2 = np.linalg.inv(X2.T @ X2) @ X2.T @ Y2
y_pred2 = X2 @ theta2
plot_multiple(X2, y2, y_pred2)
```
